### Phase 2 Preparation

- Convert `MONTH` into datetime format for proper monthly trend analysis.
- Create additional metrics such as GM%, CM%, EBITDA%, and Variance%.
- Recreate Revenue Cohort buckets as per assignment requirements.
- Create variance percentage buckets for variance-level dashboard analysis.
- Use `GROSS MARGIN` as a proxy for Contribution Margin since a separate CM column is not available.
- Use `(CITY, STORE, ZONE)` combination as the operational kitchen identifier for analysis.

```python
# Convert MONTH from string to datetime
df['MONTH'] = pd.to_datetime(df['MONTH'], format='%b-%Y')

# Confirm the conversion
print(f"MONTH dtype : {df['MONTH'].dtype}")
print(f"Min month   : {df['MONTH'].min()}")
print(f"Max month   : {df['MONTH'].max()}")
```

```python
df = df.drop(columns=['MONTH_DT'])
print(f"Columns after cleanup: {df.shape[1]}")
```

```python
df['GM %']        = (df['GROSS MARGIN']   / df['NET REVENUE']) * 100
df['CM %']        = (df['GROSS MARGIN']   / df['NET REVENUE']) * 100
df['EBITDA %']    = (df['KITCHEN EBITDA'] / df['NET REVENUE']) * 100
df['VARIANCE %']  = (df['VARIANCE']       / df['NET REVENUE']) * 100

df[['NET REVENUE', 'GM %', 'CM %', 'EBITDA %', 'VARIANCE %']].describe().round(2)
```

```python
# === Revenue cohort ===
revenue_bins   = [0, 15_00_000, 25_00_000, 35_00_000, 45_00_000, float('inf')]
revenue_labels = ['(a) Below INR 15 lacs',
                  '(b) INR 15 to 25 lacs',
                  '(c) INR 25 to 35 lacs',
                  '(d) INR 35 to 45 lacs',
                  '(e) Above INR 45 lacs']

df['REVENUE CATEGORY'] = pd.cut(df['NET REVENUE'],
                                 bins=revenue_bins,
                                 labels=revenue_labels,
                                 right=False)
```

```python
# === Variance category ===
variance_bins   = [0, 2, 3, 5, float('inf')]
variance_labels = ['(a) Var < 2%',
                   '(b) Var 2% to 3%',
                   '(c) Var 3% to 5%',
                   '(d) Var > 5%']

df['VARIANCE CATEGORY'] = pd.cut(df['VARIANCE %'],
                                  bins=variance_bins,
                                  labels=variance_labels,
                                  right=False)
```

```python
# Verify
print("Revenue Category distribution:")
print(df['REVENUE CATEGORY'].value_counts().sort_index())
print("\nVariance Category distribution:")
print(df['VARIANCE CATEGORY'].value_counts().sort_index())
```

## Phase 3: Business Insights

We move from understanding the data to drawing conclusions about the business.

### Insight 1 — Do Inactive kitchens lose money?

We saw that 47% of all kitchen-month records are marked Inactive. Are these the loss-making operations, or is the status assignment random?

```python
# Compare Active vs Inactive performance
status_comparison = df.groupby('STATUS').agg(
    avg_revenue   = ('NET REVENUE',    'mean'),
    avg_ebitda    = ('KITCHEN EBITDA', 'mean'),
    avg_ebitda_pct= ('EBITDA %',       'mean'),
    avg_variance  = ('VARIANCE %',     'mean'),
    row_count     = ('STORE',          'count')
).round(2)

status_comparison
```

Finding 1: STATUS is not correlated with profitability.

Active and Inactive kitchens show nearly identical performance:
- Avg revenue differs by ₹11K (0.3%)
- Avg EBITDA% differs by 0.4 percentage points
- Variance % is identical

This contradicts the natural assumption that "Inactive" kitchens were closed due to poor financials. Possible explanations:
- STATUS may reflect operational state (paused, under renovation, franchise transition) rather than financial health
- The field may be stale or not consistently updated
- In a real-world scenario, this would warrant clarification from the business team before using STATUS in any segmentation logic

Implication for dashboard: STATUS should be available as a filter but should not be used as a profitability indicator. Better signals: EBITDA %, EBITDA CATEGORY.

```python
# Compare cities on key profitability metrics
city_performance = df.groupby('CITY').agg(
    avg_revenue    = ('NET REVENUE',    'mean'),
    avg_ebitda     = ('KITCHEN EBITDA', 'mean'),
    avg_ebitda_pct = ('EBITDA %',       'mean'),
    avg_gm_pct     = ('GM %',           'mean'),
    row_count      = ('STORE',          'count')
).round(2).sort_values('avg_ebitda_pct', ascending=False)

city_performance
```

Finding 2: City does not meaningfully drive profitability.

```python
# Aggregate to the kitchen level using unique key
kitchen_performance = df.groupby(['CITY', 'STORE', 'ZONE MAPPING']).agg(
    total_revenue   = ('NET REVENUE',    'sum'),
    total_ebitda    = ('KITCHEN EBITDA', 'sum'),
    avg_ebitda_pct  = ('EBITDA %',       'mean'),
    months_active   = ('MONTH',          'count')
).round(2).reset_index()

# Top 10 by total EBITDA
print("=== TOP 10 KITCHENS by Total EBITDA ===")
kitchen_performance.nlargest(10, 'total_ebitda')
```

```python
# Bottom 10 by total EBITDA
print("=== BOTTOM 10 KITCHENS by Total EBITDA ===")
kitchen_performance.nsmallest(10, 'total_ebitda')
```

Finding 3: Profitability variance lives at the kitchen level - and 99% of kitchens are profitable.


```python
# Correlation between variance and profitability
correlation_vp = df['VARIANCE %'].corr(df['EBITDA %'])
correlation_vr = df['VARIANCE %'].corr(df['NET REVENUE'])
correlation_vg = df['VARIANCE %'].corr(df['GM %'])

print("=== Correlation of VARIANCE % with key metrics ===")
print(f"VARIANCE % vs EBITDA %    : {correlation_vp:.4f}")
print(f"VARIANCE % vs NET REVENUE : {correlation_vr:.4f}")
print(f"VARIANCE % vs GM %        : {correlation_vg:.4f}")
```

Finding 4: Variance (wastage) is a real driver of profitability.
